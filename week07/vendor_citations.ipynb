```python
import pandas as pd
import plotly.express as px

df=pd.read_csv('vendorlatlon.csv')

df=df[df['vend_lon']<= -118.15]
df=df[df['vend_lon']>= -118.6]
df=df[df['vend_lat']>= 33.7]
df=df[df['vend_lat']<= 34.34]
```

```python
fig = px.scatter_mapbox(df,
                        lat='vend_lat',
                        lon='vend_lon',
                        mapbox_style="stamen-toner",
                        color_discrete_sequence=["red"])

# options on the layout
fig.update_layout(
    width = 800,
    height = 800,
    title = "Interactive Vendor Citations Map 2020-2021",
    title_x=0.5 # aligns title to center
)

fig.show()
```
